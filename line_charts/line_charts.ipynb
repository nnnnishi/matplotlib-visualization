```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts.head()
```

```python
ts = ts.cumsum()
```

```python
ts.head()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
```

```python
df.head()
```

```python
df = df.cumsum()
```

```python
df.head()
```

```python
plt.figure()
df.plot()
```
