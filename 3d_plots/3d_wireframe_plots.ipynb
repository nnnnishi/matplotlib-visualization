# 3D wireframe plots


This program referred to the following site:
http://matplotlib.org/1.3.1/mpl_toolkits/mplot3d/tutorial.html

```python
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
```

```python
df = pd.DataFrame(np.random.randn(6, 8)+1*5)
```

```python
df
```

```python
row_len = df.shape[0]
col_len = df.shape[1]

x = range(1,col_len+1)
y = range(1,row_len+1)
X, Y = np.meshgrid(x, y)
Z = df.values.tolist()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('X', labelpad=0)
ax.set_ylabel('Y', labelpad=0)
ax.set_zlabel('Z', labelpad=0)
plt.title('3D wireframe plots')
```
